### Input
3. `[4:00]` Escreva um programa que pede o comprimento de dois lados de um retângulo e calcula a sua área:

```
Digite o valor do lado menor: 5
Digite o valor do lado maior: 8

A área do retângulo é 40
```

4. `[5:00]` Escreva um programa que solicita o nome e a idade de um usuário e imprime o nome e o ano de nascimento deste usuário:

```
Digite seu nome: Antonio
Digite a sua idade: 40

Antonio, você nasceu em 1979
```

5. `[5:00]` Escreva um programa que pergunta o salário de um trabalhador e o percentual de bonus ele vai receber neste mês e imprime o valor do salário com o bonus.

```
Digite o valor do salário: 1100
Digite o valor do bonus(%): 10

Seu salário esse mês é de: 1210.0
```

```
Digite um valor: 21
21 é múltiplo de 7

Digite um valor: 21
21 não é múltiplo de 7
```
